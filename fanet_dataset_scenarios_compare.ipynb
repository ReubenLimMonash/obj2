```python
import glob

dataset_path_1 = "/media/research-student/One Touch/FANET Datasets/Dataset_NP10000_MultiModulation_Hovering_NoVideo/QPSK"
dataset_path_2 = "/media/research-student/One Touch/FANET Datasets/Dataset_NP10000_MultiModulation_Hovering_NoVideo/BPSK"
scenario_list_1 = [csv.split('/')[-1][0:-11] for csv in glob.glob(dataset_path_1 + "/*GCS-Tx.csv")] # Get list of "unique" scenarios in dataset_path_1
scenario_list_2 = [csv.split('/')[-1][0:-11] for csv in glob.glob(dataset_path_2 + "/*GCS-Tx.csv")] # Get list of "unique" scenarios in dataset_path_2
# Show difference between lists
print(set(scenario_list_1) ^ set(scenario_list_2))
```
